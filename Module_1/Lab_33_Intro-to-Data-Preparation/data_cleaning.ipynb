```python
#We make checkpoint on data before moving forward

data1 = data
```

```python
# drop a column that is not useful
data_dropped = data1.drop("Department", axis=1)

data_dropped
```

```python
#we lower cases in column Profession
data_lower = data_dropped.applymap(lambda s:s.lower() if type(s) == str else s)

data_lower
```

```python
# drop columns with less than 5 NA values
data_clean = data_lower.dropna(axis=1, thresh=0)
```

```python
data_clean
```

```python
# replace na values of specific columns with mean value

data_clean["BirthYear"] = data_clean["BirthYear"].fillna(data_clean["BirthYear"].mean())
```

```python
data_clean
```

```python
# drop all rows containing null
data_drop = data_clean.dropna()

data_drop
```

```python
data_drop.groupby("Profession").nunique() 
```

```python
#We modify content of Profession to get unique job types
data_drop = data_drop.replace('etudient','student', regex=True)
data_drop.groupby("Profession").nunique() 
```
