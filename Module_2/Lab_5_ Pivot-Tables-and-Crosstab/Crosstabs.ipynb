### 1. Import the moovies dataset 

# https://www.kaggle.com/carolzhangdc/imdb-5000-movie-dataset

```python
from google.colab import files
uploaded = files.upload()
```

```python
import io
df_raw = pd.read_csv(io.BytesIO(uploaded['movie_metadata.csv']))
# Dataset is now stored in a Pandas Dataframe

df_raw
```

### 2. Use crosstabs to show the number of moovies per country

```python
ct1 = pd.crosstab(df_raw.country,df_raw.director_name, margins=True, margins_name='Total')
ct1['Total']
```

### 3. Create a crosstab to show the number of films per director and genre splitted by country

```python
ct2 = pd.crosstab([df_raw.director_name, df_raw.genres],[df_raw.country], dropna=False)
ct2
```

### 4. Average duration of filmes grouped by color

```python
ct3 = pd.crosstab(df_raw.color, df_raw.movie_title, values=df_raw.duration,aggfunc='mean',margins=True, margins_name='Total').round(0)
ct3['Total']
```

### 5. Average number of likes per director (assuming that total film's likes are counted as a sum of actor's likes)

```python
ct4 = pd.crosstab(df_raw.director_name, df_raw.movie_title, values=df_raw.movie_facebook_likes,aggfunc='mean',margins=True, margins_name='Total').round(0)
ct4['Total']
```
