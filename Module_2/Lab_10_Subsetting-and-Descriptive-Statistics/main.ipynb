#### Explore your dataset by looking at the data types and summary statistics. Comment your results.

```python
orders.dtypes
```

```python
"""
your comments here
7 integer cols, in which amount_spent, unitprice need to be changed into float dtype
"""
#orders.apply(lambda x: x.str.replace(',','.'))

order = orders.replace(',','.',regex=True)

order[['UnitPrice', 'amount_spent']] = order[['UnitPrice', 'amount_spent']].astype(np.float64)
```

####  What is the average purchase price?

```python
# your code here
order['amount_spent'].mean()
```

#### What are the highest and lowest purchase prices? 

```python
# your code here
p_min = order['amount_spent'].min()
p_max = order['amount_spent'].max()

print('lowest purchase price is: ',p_min, '\nhighest purchase price is: ',p_max)
```
